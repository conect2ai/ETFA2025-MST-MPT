## üöó Freematics Output Analysis

This section analyzes the data collected from the Freematics device during real-time execution.  
The results are compared to the original input data to assess the effectiveness of the deployed algorithms.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df_fastback = pd.read_csv("../data/data_mstedarls_mptedarls_fastback.csv")
df_polo = pd.read_csv("../data/data_mstedarls_mptedarls_polo.csv")

# Rename columns if necessary
df_fastback = df_fastback.rename(columns={"time_mstedarls_us": "MST", "time_mptedarls_us": "MPT"})
df_polo = df_polo.rename(columns={"time_mstedarls_us": "MST", "time_mptedarls_us": "MPT"})

# Restructure to long format
df_fastback_long = df_fastback[["MST", "MPT"]].melt(var_name="Algorithm", value_name="Execution Time (¬µs)")
df_fastback_long["Vehicle"] = "Fastback"

df_polo_long = df_polo[["MST", "MPT"]].melt(var_name="Algorithm", value_name="Execution Time (¬µs)")
df_polo_long["Vehicle"] = "Polo"

df_all = pd.concat([df_fastback_long, df_polo_long], ignore_index=True)
```

## üîó Outlier Correlation Analysis

This section examines the correlation between the detected outliers across different features.  
Correlation matrices are computed and visualized to assess whether the detection patterns exhibit dependencies or shared anomalies among the time series.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ======================
# Function to save heatmap
# ======================
def save_heatmap(df_corr, output_path, cmap):
    plt.figure(figsize=(6, 5))
    sns.heatmap(
        df_corr,
        annot=True,
        cmap=cmap,
        vmin=0,
        vmax=1,
        square=True,
        cbar=True,
        annot_kws={"size": 16}
    )
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    plt.tight_layout()
    plt.savefig(output_path, dpi=300)
    plt.close()

# ======================
# Load data
# ======================
df_fastback = pd.read_csv("../data/output_fastback_mstedarls.csv")[[
    "outlier_speed", "outlier_rpm", "outlier_tp", "outlier_load", "outlier_timing"
]]

df_polo = pd.read_csv("../data/output_polo_mstedarls.csv")[[
    "outlier_speed", "outlier_rpm", "outlier_tp", "outlier_load", "outlier_timing"
]]

# Rename columns for better visualization
df_fastback.columns = ["Speed", "RPM", "TP", "Load", "Timing"]
df_polo.columns = ["Speed", "RPM", "TP", "Load", "Timing"]

# ======================
# Generate and save heatmaps
# ======================
save_heatmap(
    df_fastback.corr(),
    "../figures/heatmap_corr_fastback_mstedarls.png",
    cmap="Blues"
)

save_heatmap(
    df_polo.corr(),
    "../figures/heatmap_corr_polo_mstedarls.png",
    cmap="Greens"
)
```

## üìä Correction Ratio Evaluation

The proportion of samples corrected by each algorithm is analyzed.  
This metric helps to quantify the aggressiveness and sensitivity of the correction strategies applied.



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ==========================
# Load CSV files
# ==========================
df_fastback = pd.read_csv("../data/data_mstedarls_mptedarls_fastback.csv")
df_polo = pd.read_csv("../data/data_mstedarls_mptedarls_polo.csv")

# ==========================
# Define flag columns
# ==========================
cols_mst = [
    "mstedarls_flag_speed",
    "mstedarls_flag_rpm",
    "mstedarls_flag_tp",
    "mstedarls_flag_load",
    "mstedarls_flag_timing"
]

cols_mpt = [
    "mptedarls_speed",
    "mptedarls_rpm",
    "mptedarls_tp",
    "mptedarls_load",
    "mptedarls_timing",
    "mptedarls_flag"  # used to count if a correction was applied
]

# ==========================
# Build correction summary
# ==========================
correction_summary = pd.DataFrame({
    "Vehicle": ["Fiat Fastback", "Fiat Fastback", "VW Polo", "VW Polo"],
    "Algorithm": ["MST", "MPT", "MST", "MPT"],
    "Total Corrections": [
        df_fastback[cols_mst].values.sum(),
        df_fastback["mptedarls_flag"].sum(),
        df_polo[cols_mst].values.sum(),
        df_polo["mptedarls_flag"].sum()
    ],
    "Total Processed Variables": [
        df_fastback[cols_mst].shape[0] * len(cols_mst),
        df_fastback[cols_mpt].shape[0],
        df_polo[cols_mst].shape[0] * len(cols_mst),
        df_polo[cols_mpt].shape[0]
    ]
})

correction_summary["Correction Proportion (%)"] = (
    correction_summary["Total Corrections"] / correction_summary["Total Processed Variables"]
) * 100

# ==========================
# Generate bar chart
# ==========================
vehicles = ["Fiat Fastback", "VW Polo"]
x = np.arange(len(vehicles))
bar_width = 0.35

mst_vals = correction_summary[correction_summary["Algorithm"] == "MST"]["Correction Proportion (%)"].values
mpt_vals = correction_summary[correction_summary["Algorithm"] == "MPT"]["Correction Proportion (%)"].values

fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(x - bar_width/2, mst_vals, width=bar_width, label="MST", color="#76b5c5")
ax.bar(x + bar_width/2, mpt_vals, width=bar_width, label="MPT", color="#f4a582")

ax.set_ylabel("Correction Proportion (%)", fontsize=18)
# ax.set_title("Corrections relative to total processed variables")
ax.set_xticks(x)
ax.set_xticklabels(vehicles, fontsize=16)
ax.set_ylim(0, max(max(mst_vals), max(mpt_vals)) + 1)
ax.legend(fontsize=18)
ax.grid(True, axis="y", linestyle="--", alpha=0.6)

plt.tight_layout()
plt.savefig("../figures/grafico_correcao_mst_mpt.png", dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ================================
# Load CSV files
# ================================
original_fastback = pd.read_csv("../data/dados_sem_outliers_fastback.csv")
corrected_mpt_fastback = pd.read_csv("../data/output_fastback_sem_outlier.csv")
corrected_mst_fastback = pd.read_csv("../data/output_fastback_mstedarls.csv")

original_polo = pd.read_csv("../data/dados_sem_outliers_polo.csv")
corrected_mpt_polo = pd.read_csv("../data/output_polo_sem_outlier.csv")
corrected_mst_polo = pd.read_csv("../data/output_polo_mstedarls.csv")

# ================================
# Features and readable names
# ================================
features = ["speed", "rpm", "tp", "load", "timing"]
readable_names = {
    "speed": "Speed\n(km/h)",
    "rpm": "RPM\n(rpm)",
    "tp": "Throttle\nPosition (%)",
    "load": "Engine\nLoad (%)",
    "timing": "Timing\nAdvance (¬∞)"
}

# ================================
# Function to generate figures
# ================================
def plot_time_series(original, corrected_mpt, corrected_mst, filename):
    fig, axs = plt.subplots(len(features), 1, figsize=(12, 10), sharex=True)
    
    for i, feature in enumerate(features):
        axs[i].plot(original[feature], label="Original", color="black", linewidth=1)
        axs[i].plot(corrected_mpt[f"y_{feature}"], label="MPT", color="#f4a582", linewidth=0.9)
        axs[i].plot(corrected_mst[f"{feature}_corrected"], label="MST", color="#76b5c5", linewidth=0.9)
        axs[i].set_ylabel(readable_names[feature], fontsize=18)  # Fonte maior para o eixo Y
        axs[i].grid(True, linestyle="--", alpha=0.4)
        if i == 0:
            axs[i].legend(loc="upper right", fontsize=16)  # Fonte maior para a legenda

    axs[-1].set_xlabel("Time (samples)", fontsize=18)  # Fonte maior para o eixo X
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.savefig(f"../figures/{filename}", dpi=300)
    plt.close()

# ================================
# Generate figures for both vehicles
# ================================
plot_time_series(
    original_fastback,
    corrected_mpt_fastback,
    corrected_mst_fastback,
    "time_series_fastback.png"
)

plot_time_series(
    original_polo,
    corrected_mpt_polo,
    corrected_mst_polo,
    "time_series_polo.png"
)
```

## üìâ Probability Density Function (PDF) Analysis

Probability Density Functions (PDFs) of the corrected series are compared to those of the original clean series.  
The closeness of the PDFs indicates the ability of each algorithm to reconstruct the statistical properties of the original data.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ===============================
# Load data
# ===============================
orig_fast = pd.read_csv("../data/dados_sem_outliers_fastback.csv")
corr_mpt_fast = pd.read_csv("../data/output_fastback_sem_outlier.csv")
corr_mst_fast = pd.read_csv("../data/output_fastback_mstedarls.csv")

orig_polo = pd.read_csv("../data/dados_sem_outliers_polo.csv")
corr_mpt_polo = pd.read_csv("../data/output_polo_sem_outlier.csv")
corr_mst_polo = pd.read_csv("../data/output_polo_mstedarls.csv")

# ===============================
# Variables and readable names (with line breaks)
# ===============================
variables = ["speed", "rpm", "tp", "load", "timing"]
readable_names = {
    "speed": "Speed (km/h)",
    "rpm": "RPM (rpm)",
    "tp": "Throttle Position (%)",
    "load": "Engine Load (%)",
    "timing": "Timing Advance (¬∞)"
}

# ===============================
# Function to plot density PDFs
# ===============================
def plot_pdfs(orig, mpt, mst, vehicle_name, file_name):
    fig, axs = plt.subplots(5, 1, figsize=(10, 12), sharex=False)

    for i, var in enumerate(variables):
        sns.kdeplot(orig[var], label="Original", color="black", lw=1.8, ax=axs[i])
        sns.kdeplot(mpt[f"y_{var}"], label="MPT", color="#f4a582", lw=1.5, ax=axs[i])
        sns.kdeplot(mst[f"{var}_corrected"], label="MST", color="#76b5c5", lw=1.5, ax=axs[i])

        # Y-axis: always "Density"
        axs[i].set_ylabel("Density", fontsize=18, labelpad=12)

        # X-axis: specific feature name
        axs[i].set_xlabel(readable_names[var], fontsize=18, labelpad=10)

        axs[i].tick_params(axis='both', labelsize=18)
        axs[i].grid(True, linestyle="--", alpha=0.4)

        if i == 0:
            axs[i].legend(loc="upper right", fontsize=15)

    fig.align_ylabels(axs)
    plt.subplots_adjust(left=0.12, hspace=0.9)
    plt.savefig(f"../figures/{file_name}", dpi=300)
    plt.close()

# ===============================
# Generate plots
# ===============================
plot_pdfs(orig_fast, corr_mpt_fast, corr_mst_fast,
          "Fiat Fastback", "pdfs_fastback.png")

plot_pdfs(orig_polo, corr_mpt_polo, corr_mst_polo,
          "Volkswagen Polo", "pdfs_polo.png")
```
